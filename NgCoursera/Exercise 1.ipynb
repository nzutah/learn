```python
data = np.loadtxt('./data/ex1data1.txt', delimiter=',')

# X is all of the training data, with a column of ones inserted
X = np.column_stack((np.ones_like(data[:,0]), data))
N = X.shape[1] - 1
```

### 2.2 Gradient Descent

```python
def gradient_descent(fun, x, y, theta, alpha=.01, maxiter=1000, tol=1.e-6, disp=0):
    '''Find

    fun : callable
        The cost function
    x : ndarray (m,n)
        x[i,j] is the jth feature of example i
    y : ndarray (m,)
        y[i] is the ith training examples
    theta : ndarray(n,)
        initial guess of coefficients

    '''
    m = y.shape[0] # number of training examples
    assert x.shape[1] == theta.shape[0]
    dnom = abs(average(y))
    J = []
    for i in range(maxiter):
        d = dot(x, theta) - y
        dtheta = alpha / m * dot(d.T, x)
        theta -= dtheta
        J.append(fun(x, y, theta))
        if norm(dtheta) / dnom < tol:
            break
    else:
        warnings.warn('maximum gradient descent iterations reached', RuntimeWarning)

    if not disp:
        return theta

    return theta, array(J)
```

```python
theta = np.zeros(N) # initialize fitting parameters

def costfunc(x, y, theta):
    d = dot(x, theta) - y
    return dot(d, d) / 2. / x.shape[0]

# compute and display initial cost
costfunc(X[:,:N], X[:,-1], theta)

# run gradient descent
theta, Jh = gradient_descent(costfunc, X[:,:N], X[:,-1], 
                             theta, tol=1e-3, disp=1)
theta
```

```python
# Plot the linear fit
plot(X[:, 1], X[:,-1], 'rx', markersize=10, label='Training data')
ylabel('Profit in $10,000s')
xlabel('Population of City in 10,000s');
plot(X[:, 1], np.dot(X[:, :N], theta), label='Linear regression')
legend(loc='best');
```

```python
# Predict values for population sizes of 35,000 and 70,000
p = np.dot([1, 3.5], theta)
p*10000
```

### 2.4 Visualizing $J(\theta_0, \theta_1)$

```python
# Grid over which we will calculate J
m, n = 100, 100
T0 = np.linspace(-10, 10, m)
T1 = np.linspace(-1, 4, n)
J = np.array([costfunc(X[:,:N], X[:,-1], [T0[i], T1[j]])
              for i in range(m) for j in range(n)]).reshape(m, n)

# Surface plot
fig = figure()
ax = fig.add_subplot(111, projection='3d')
xx, yy = np.meshgrid(T0, T1)
ax.plot_surface(xx, yy, J, rstride=1, cstride=1, cmap=cm.coolwarm);
```

## 3. Linear Regression with Multiple Variables

### 3.1 Feature Normalization

```python
data = np.loadtxt('./data/ex1data2.txt', delimiter=',')
X = np.column_stack((np.ones_like(data[:,0]), data))

plot(X[:, 2], X[:, 1], 'x')
ylabel('Square feet')
xlabel('Number of bedrooms');
```

### 3.2 Gradient Descent

```python
def normalize(x):
    '''Normalize each feature

    Parameters
    ----------
    x : ndarray (m,n)
        x[i,j] is the jth feature of example i

    Returns
    -------
    xb : ndarray (m,n)
        Normalized x
    mu, sigma : ndarray (m,)
        mean value and standard deviation

    Notes
    -----
    For each feature dimension, the mean of the feature is computed and
    subtracted from the dataset, storing the mean value in mu. Next, the
    standard deviation of each feature is computed and each feature by it's
    standard deviation, storing the standard deviation in sigma.

    '''
    xb = x.copy()
    m, n = x.shape
    mu, sigma = zeros(n), zeros(n)
    for i in range(n):
        mu[i] = mean(xb[:, i])
        sigma[i] = std(xb[:, i])
        if abs(sigma[i]) < 1.e-12:
            continue
        xb[:, i] = (xb[:, i] - mu[i]) / sigma[i]
    return xb, mu, sigma
```

```python
Xb, mu, sigma = normalize(X)
N = Xb.shape[1] - 1

for alpha in (0.3, 0.1, 0.03, 0.01):
    theta = np.zeros(N) # initialize fitting parameters
    theta, Jh = gradient_descent(costfunc, Xb[:,:N], Xb[:,-1], theta, 
                                 alpha=alpha, tol=1e-3, disp=1)
    plot(range(1, 51), Jh[:50], label='alpha={0}'.format(alpha))
xlabel('Number of iterations')
ylabel('Cost J')
legend();
```

### 3.3 Normal Equation

```python
def least_squares(x, y):
    z = dot(x.T, y)
    return dot(inv(dot(x.T, x)), z)
normal_eqn = least_squares
```

```python
theta = normal_eqn(X[:, :N], X[:, -1])
theta
```
