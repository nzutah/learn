### 1.2 Implementation

#### 1.2.1 Sigmoid Function

```python
def sigmoid(z):
    return 1. / (1. + exp(-z))
```

#### 1.2.2 Cost Function

```python
def costfun(theta, x, y, lam=None):
    # the cost function
    x, y = asarray(x), asarray(y)
    z = dot(x, theta)
    ht = sigmoid(z)
    m, n = x.shape
    ht[where(ht >= 1.)] = .99
    J = (-dot(y, log(ht)) - dot((1 - y), log(1 - ht))) / m
    if lam is not None:
        J += lam / 2. / m * dot(theta[1:], theta[1:])
    return J

def costfun_grad(theta, x, y, lam=None):
    # the cost function gradient
    x, y = asarray(x), asarray(y)
    z = dot(x, theta)
    ht = sigmoid(z)
    m, n = x.shape
    grad = array([dot(ht - y, x[:, j]) for j in range(n)]) / m
    if lam is not None:
        grad[1:] += lam / m * theta[1:]
    return grad

def predict(theta, x):
    return round(sigmoid(dot(x, theta)))
    
theta = zeros(nfeatures)
J = costfun(theta, data.drop('Admit', axis=1), data['Admit'])
assert abs(J - .693) < 5.e-4
```

#### 1.2.3 Learning Parameters

```python
theta = fmin_cg(costfun, theta, disp=0,
                args=(data.drop('Admit', axis=1), data['Admit']),
                fprime=costfun_grad)
J = costfun(theta, data.drop('Admit', axis=1), data['Admit'])
assert abs(J - .203) < 5e-3
```

```python
plot(admitted['Test 1'], admitted['Test 2'], 'k+', 
     label='Admitted')
plot(not_admitted['Test 1'], not_admitted['Test 2'], 'yo', 
     label='Not Admitted')

# Only need 2 points to define a line, so choose two endpoints
xp = array([min(data['Test 1']),  max(data['Test 1'])])

# Calculate the decision boundary line
yp = (-1. / theta[2]) * (theta[1] * xp + theta[0])

# plot the decision boundary
plot(xp, yp)
xlim(30, 100)
ylim(30, 100)
legend();
```

#### 1.2.4 Evaluating Logistic Expression

```python
assert abs(sigmoid(dot(theta, [1, 45, 85])) - .776) < 5e-3
```

```python
def warn(message, *args):
    message = message % args
    print '*** warning: %s' % message
for (i, row) in data.iterrows():
    prediction = predict(theta, [1, row['Test 1'], row['Test 2']])
    if prediction != row['Admit']:
        T1, T2 = row['Test 1'], row['Test 2']
        warn('incorrect prediction for '
             'Test 1 = %.2f, Test 2 = %.2f, ', T1, T2)
```

## 2. Regularized Logistic Regression

### 2.1 Visualizing the Data

```python
data = pandas.read_csv('data/ex2data2.txt', header=None,
                       names=('Test 1', 'Test 2', 'Admit'))
y1 = data.loc[data['Admit'] == 1]
y0 = data.loc[data['Admit'] == 0]
plot(y1['Test 1'], y1['Test 2'], 'k+', label='y=1')
plot(y0['Test 1'], y0['Test 2'], 'yo', label='y=0')
legend();
```

### 2.2 Feature Mapping

```python
def map_feature(x1, x2):
    x1 = asarray(x1)
    x2 = asarray(x2)
    data = pandas.DataFrame()

    data['X1'] = x1
    data['X2'] = x2
    degree = 6
    for i in range(1, degree + 1):
        for j in range(i+1):
            key = '{0}{1}'.format(i, j)
            data[key] = x1 ** (i - j) * x2 ** j
            
    # Add the intercept
    data.insert(0, 'Intercept', 1.)
    return data
X = map_feature(data['Test 1'], data['Test 2'])
nfeatures = X.shape[1]
```

### 2.3 Cost Function and Gradient

```python
theta = zeros(nfeatures)
lam = 1.
J = costfun(theta, X, data['Admit'])
assert abs(J - .693) < 5e-4
```

#### 2.3.1 Learning Parameters

```python
theta = fmin_cg(costfun, theta, disp=0, args=(X, data['Admit'], lam),
                fprime=costfun_grad)
J = costfun(theta, X, data['Admit'])
```

### 2.4 Plotting the Decision Boundary

```python
# The grid range
n = 50
u = linspace(-1, 1.5, n)
v = linspace(-1, 1.5, n)

z = zeros((n, n))
# Evaluate z = theta.x over the grid
for i in range(n):
    for j in range(n):
        z[i,j] = dot(map_feature(u[i:i+1], v[j:j+1]), theta)

y1 = data.loc[data['Admit'] == 1]
y0 = data.loc[data['Admit'] == 0]
plot(y1['Test 1'], y1['Test 2'], 'k+', label='y=1')
plot(y0['Test 1'], y0['Test 2'], 'yo', label='y=0')

# Plot z = 0
# Notice you need to specify the range [0, 0]
contour(u, v, z, [0, 0], label='Decision Boundary')

legend();
```
